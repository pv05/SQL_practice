### Not like with where

```python
%%sql 
# Not starts with "ch"
select * from Products where ProductName not like "ch%"; 
```

```python
%%sql 
# ends with "ing"
select * from Products where ProductName not like "%ing"; 
```

```python
%%sql 
# contains "th"  
select * from Products where ProductName not like "%th%";  
```

```python
%%sql 
# like with where and having
select * from Products where ProductName like "ch%" having Price < 20;
```

```python
%%sql 
# using AND | # alternative method of above command
select * from Products where ProductName like "ch%" and Price < 20; 
```

```python
%%sql 
# where,or,and,like 
select * from Products where ( Price%6 = 0 or Price%10 = 0 ) and ProductName like "%s";
```

```python
%%sql 
select * from Products where ( Price%6 = 0 and Price%10 = 0 ) and ProductName like "%s";
```

## Date query

```python
%%sql 
select * from orders where OrderDate between '1996-07-10' and '1996-07-31' 
```

```python
%%sql 
# Get the current date and time:
select now() 
```

```python
%%sql 
# Get the current date:
select curdate() 
```

```python
%%sql 
# Get the current time:
select curtime() 
```

```python
%%sql 
# Extract year, month or day from a date:
select year(orderdate) , month(orderdate) , day(orderdate) from orders 
```

```python
%%sql 
# Calculate the difference between two dates from todays
SELECT DATEDIFF(curdate() , orderdate) FROM orders;
```

```python
%%sql 
#Get the last day of the month:
SELECT LAST_DAY(orderdate) FROM orders;
```

```python
%%sql 
#Get the week number of a date:
SELECT WEEK(orderdate) FROM orders;
```

# Table Joins

```python
%%sql 
select * from df1  
```

```python
%%sql 
select * from df2 
```

```python
%%sql 
# INNER
select * from df1 inner join df2 on df1.id = df2.id  
```

```python
%%sql 
# LEFT
select * from df1 left join df2 on df1.id = df2.id 
```

```python
%%sql 
# RIGHT
select * from df1 right join df2 on df1.id = df2.id 
```

```python
%%sql  
# full join not working directly but, we gona do with union and left and right join
select * from df1 left join df2 on df1.id =df2.id 
union
select * from df1 right join df2 on df1.id = df2.id 
```

```python
%%sql
# SELF join
select * from selfjoindf s1 left join selfjoindf s2 on s1.id = s2.manager_id 
```

```python
%%sql 
select * from crossdf1 
```

```python
%%sql 
select * from crossdf2 
```

```python
%%sql 
#CROSS Join
select * from crossdf1 cross join crossdf2 
```

## Group  BY

```python
%%sql 
select s.shippername, count(orderid) from orders o left join shippers s on o.ShipperID = s.ShipperID
group by ShipperName 
```

```python
%%sql 
SELECT country , COUNT(CustomerID) as "total customer" from customers
GROUP BY Country
ORDER BY COUNT(CustomerID) DESC;
```
